```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from glob import glob

%matplotlib inline

# Autor: Enzo Gaggioli, e.l.gaggioli@gmail.com.
# IFIBA-CONICET-UBA

# Create a "key" to make sure that the files are read in numerical order
import re
numbers = re.compile(r'(\d+)')
def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts

# Create empty arrays to store the means and standard deviations
t_values = []
dt_values = []

directory = '/home/enzo/Desktop/Enzo/25-04/'
#directory = '/home/enzo/Escritorio/Dedalus_work/Experimentos/25-04/'

# Read the TIFF files for different concentrations of milk
for concentration in ['leche_10.tif.frames', 'leche_20.tif.frames',
                      'leche_40.tif.frames', 'leche_60.tif.frames', 'leche_80.tif.frames', 'leche_100.tif.frames']:
    # Create a list to store the array for each frame
    imarray_list = []
    # Read each frame and convert it to a numpy array
    for filename in sorted(glob(directory + concentration + '/*.tif'), key=numericalSort):
        im = Image.open(filename)
        imarray = np.array(im)
        imarray_list.append(imarray)
    # Convert the list of arrays to a numpy array
    imarray_all = np.asarray(imarray_list)
    # Calculate the mean and standard deviation of the pixel values
    t = np.mean(imarray_all)
    dt = np.std(imarray_all)
    # Print the results
    print('I({}%) = {:.2f}, std = {:.2f}'.format(concentration.split('_')[-1].split('.')[0], t, dt))
    # Store the results in the arrays
    t_values.append(t)
    dt_values.append(dt)

# Plot the mean and standard deviation as a function of concentration
#c = [0, 10, 20, 40, 60, 80, 100]
#plt.plot(c, t_values, 'o-', label='Mean')
#plt.plot(c, dt_values, 'o-', label='Standard Deviation')
#plt.xlabel('Concentration (%)')
#plt.ylabel('Pixel Value')
#plt.legend()
#plt.show()

sum_t = sum(t_values)

#print('sum_t=',sum_t)

t_values = t_values/sum_t
dt_values = dt_values/sum_t

x_values = [10, 20, 40, 60, 80, 100]  # These are the concentration values in %
y_values = t_values
y_err_values = dt_values

plt.errorbar(x_values, y_values, yerr=y_err_values, fmt='o')
plt.xlabel('Concentration (%)')
plt.ylabel('Mean intensity')
plt.title('Mean intensity vs concentration')
plt.show()
```
